```python
import os
from polly.auth import Polly
AUTH_TOKEN=(os.environ['POLLY_REFRESH_TOKEN']) # Obtain authentication tokens
Polly.auth(AUTH_TOKEN)
from polly.omixatlas import OmixAtlas
omixatlas = OmixAtlas()
```

```python
# testing point 1 -> passed
# table names passed 
repo_key_1 = "geo"
resp = omixatlas.get_schema(repo_key_1, ["datasets", "samples"], source="all", data_type="all")
print("-----dataset payload------")
print(resp.datasets)
print("\n")
print("-----sample payload-----")
print(resp.samples)
```

```python
# testing point 2 -> passed
# valid schema level given as param
repo_key_1 = "geo"
resp = omixatlas.get_schema(repo_key_1, ["dataset"], source="all", data_type="all")
print("-----dataset payload------")
print(resp.dataset)
```
