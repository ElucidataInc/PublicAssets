```python
from polly.auth import Polly

from polly.polly_kg import PollyKG

AUTH_KEY = ""
Polly.auth(AUTH_KEY, env="testpolly")

kg = PollyKG()
```

```python
res1 = kg.get_graph_summary()

print(res1)
```

```python
res2 = kg.get_engine_status()

print(res2)
```

```python
# running gremlin query
res3 = kg.run_gremlin_query("g.V().limit(5).project('id').by(id).by(valueMap())")
print(res3)
```

```python
# running gremlin query with id and property
res3 = kg.run_gremlin_query("g.V().limit(5).project('id', 'property').by(id).by(valueMap())")
print(res3)
```

```python
# running gremlin query timeout
res3 = kg.run_gremlin_query("g.V().limit(5000).project('id', 'property').by(id).by(valueMap())")
print(res3)
```

```python
# running gremlin query wrong query
res3 = kg.run_gremlin_query("g.V().liit(5000).project('id', 'property').by(id).by(valueMap())")
print(res3)
```

```python
# running gremlin query empty query
res3 = kg.run_gremlin_query("")
print(res3)
```

```python
# running gremlin query none query
res3 = kg.run_gremlin_query()
print(res3)
```

```python
#running opencypher query
res4 = kg.run_opencypher_query("MATCH (node) RETURN node LIMIT 1")
print(res4)
```

```python
#running opencypher query wrong query
res4 = kg.run_opencypher_query("MATCH (node) RETRN node LIMIT 1")
print(res4)
```

```python
#running opencypher query empty query
res4 = kg.run_opencypher_query("")
print(res4)
```
