```python
import os
from polly.auth import Polly
AUTH_TOKEN=(os.environ['POLLY_REFRESH_TOKEN']) # Obtain authentication tokens
Polly.auth(AUTH_TOKEN)
from polly.omixatlas import OmixAtlas
omixatlas = OmixAtlas()
```

```python
from polly_validator.validators import dataset_metadata_validator
```

```python
from polly_validator.validators import sample_metadata_validator
```

```python
dataset_id = "GSE100035_GPL21103"
repo_key = "geo"
file_name = f"{dataset_id}.gct"
data = omixatlas.download_data(repo_key, dataset_id)
url = data.get('data').get('attributes').get('download_url')
status = os.system(f"wget -O '{file_name}' '{url}'")
if status == 0:
    print("Downloaded data successfully")
else:
    raise Exception("Download not successful")
```

```python
import pandas as pd
import cmapPy
from cmapPy.pandasGEXpress.parse_gct import parse

gct_obj = parse(file_name) # Parse the file to create a gct object
df_real = gct_obj.data_df # Extract the dataframe from the gct object
col_metadata = gct_obj.col_metadata_df 
```

```python
col_metadata
```

```python
query = 'DESCRIBE geo.samples'
schema_df_sample = omixatlas.query_metadata(query, query_api_version="v2")
```

```python
err_sample, status_sample = sample_metadata_validator.check_metadata_for_errors(repo='geo',
                                                                              schema_df=schema_df_sample,
                                                                              metadata_list=col_metadata.to_dict('records'),
                                                                              validate_on='value',
                                                                              env="testpolly",
                                                                              auth_token=AUTH_TOKEN,
                                                                                print_table=True)
```
