### Checking for schema of a repository

```python
query = "SHOW TABLES IN geo"
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Querying for schema of dataset level metadata of GEO

```python
query = "DESCRIBE geo.datasets"
#query = "DESCRIBE geo.samples"
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Querying for dataset-id, description, number of samples etc. from GEO where platform is 'RNASeq', number of samples >20.

```python
query = """SELECT dataset_id, description, total_num_samples, disease, tissue FROM geo.datasets 
            WHERE platform = 'RNASeq' AND total_num_samples > 20 ORDER BY total_num_samples desc"""
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Querying for datasets from GEO where disease array has any item consisting of keyword 'neoplasm'

```python
#Querying for datasets from GEO where disease array has any item consisting of keyword 'neoplasm'
query = "SELECT disease FROM geo.datasets WHERE CARDINALITY(FILTER(disease, d -> LOWER(d) like '%neoplasm%')) > 0"
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Querying for datasets and number of samples in a dataset from GEO where disease array has an exact term 'Obesity'

```python
query = "SELECT dataset_id, total_num_samples FROM geo.datasets WHERE CONTAINS(disease, 'Obesity')"
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Get the abundance of different tissues across all datasets in GEO

```python
query = """SELECT tissue_name, COUNT((tissue_name)) AS tissue_count FROM geo.datasets 
            CROSS JOIN UNNEST(tissue) AS t(tissue_name) GROUP BY tissue_name"""
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Fetch the list of all the dataset IDs where expression value of gene 'brd4' is reported

```python
query = "SELECT src_dataset_id, name FROM geo.features WHERE LOWER(name) = 'brd4'"
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Get the name of samples, dataset ID and extract_protocol_ch1 where spectroscopy is mentioned in the extract_protocol_ch1

```python
query = """SELECT name, src_dataset_id, extract_protocol_ch1 FROM geo.samples
         WHERE LOWER(extract_protocol_ch1) LIKE '%spectroscopy%'"""
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Get the name of gene and number of samples where information about particular gene is curated 

```python
query = """SELECT gene, COUNT(*) AS count FROM geo.samples GROUP BY gene ORDER BY count DESC """
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Get number of datasets with all combinations of disease and tissues in GEO

```python
query = """SELECT tissue_name, disease_name, COUNT(*) AS COUNT_COL FROM geo.datasets 
    CROSS JOIN UNNEST(tissue) AS t(tissue_name) 
    CROSS JOIN UNNEST (disease) AS t(disease_name)
    GROUP BY tissue_name, disease_name
    ORDER BY COUNT_COL desc"""
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Get dataset_IDs where disease is obesity and organism can be either human or rat

```python
query = """SELECT dataset_id, organism_name, disease_name FROM geo.datasets 
    CROSS JOIN UNNEST(organism) AS t(organism_name) 
    CROSS JOIN UNNEST (disease) AS t(disease_name) 
    WHERE LOWER (disease_name) LIKE 'obesity' AND (LOWER(organism_name) LIKE 'homo sapiens' OR LOWER (organism_name) LIKE 'rattus norvegicus')"""
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

## JOIN Queries

### Get microarray datasets and their samples where the ‘HeLa’ cell line was used on a microarray platform

```python
#Get microarray datasets and their samples where the ‘HeLa’ cell line was used on a microarray platform

query = """  SELECT datasets.dataset_id AS dataset_id, sample_name AS sample_name FROM geo.datasets AS datasets
    JOIN (
        SELECT src_dataset_id AS dataset_id, name AS sample_name FROM geo.samples WHERE CONTAINS(cell_line, 'HeLa')
         ) AS hela_samples
    ON datasets.dataset_id = hela_samples.dataset_id WHERE LOWER(platform) = 'microarray'
    ORDER BY datasets.dataset_id"""
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Queries using a sub-query

```python
query = """SELECT src_dataset_id AS dataset_id, name AS sample_name FROM geo.samples
    WHERE CONTAINS (cell_line, 'HeLa') AND src_dataset_id 
    IN (SELECT dataset_id FROM geo.datasets WHERE LOWER (platform) = 'microarray')"""
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```
